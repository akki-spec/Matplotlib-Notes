```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#bar() -  compare categories of data by representing each category with a bar

categories = np.array(['Grains','Fruit','Vegetables','Protein','Dairy','Sweets'])
value = np.array([5,3,8,9,7,2])

plt.bar(categories,value)
plt.show()
```

```python


categories = np.array(['Grains','Fruit','Vegetables','Protein','Dairy','Sweets'])
value = np.array([5,3,8,9,7,2])

plt.bar(categories,value, color = "skyblue")
plt.title("Daily Nutritional Value")
plt.xlabel("Categories")
plt.ylabel("Value")
plt.show()
```

```python


categories = np.array(['Grains','Fruit','Vegetables','Protein','Dairy','Sweets'])
value = np.array([5,3,8,9,7,2])

plt.barh(categories,value, color = "skyblue")
plt.title("Daily Nutritional Value")
plt.xlabel("Categories")
plt.ylabel("Value")
plt.show()
```
