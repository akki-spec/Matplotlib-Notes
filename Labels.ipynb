```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([2010, 2015, 2020, 2025])

y1 = np.array([20, 40, 100, 80])
y2 = np.array([22, 30, 95, 100])
y3 = np.array([30, 70, 80, 3])

plt.title("class size", fontsize=25,
          family="serif",
          fontweight = "bold",
          color = "#2b11d6")

plt.xlabel("year",fontsize=15,
          family="serif",
          fontweight = "bold",
          color = "#2b11d6")

plt.ylabel("class size",fontsize=15,family="serif",
          fontweight = "bold",
          color = "#2b11d6")

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([2010, 2015, 2020, 2025])

y1 = np.array([20, 40, 100, 80])
y2 = np.array([22, 30, 95, 100])
y3 = np.array([30, 70, 80, 3])

plt.title("class size", fontsize=25,
          family="serif",
          fontweight = "bold",
          color = "#2b11d6")

plt.xlabel("year",fontsize=15,
          family="serif",
          fontweight = "bold",
          color = "#2b11d6")

plt.ylabel("class size",fontsize=15,family="serif",
          fontweight = "bold",
          color = "#2b11d6")

plt.tick_params(axis="both")# axis - x,y,both and also add color and labelsize

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.xticks(x)
plt.show()
```
