```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4,5]
y = [5,10,15,20,25]

#grid() =  helps makes plot easier by adding reference line
plt.grid()

plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [5,10,15,20,25]

plt.grid( axis="y",
         color = "lightgrey",
         linewidth = 2,
         linestyle = "dashed")

plt.plot(x,y)
plt.show()
```
