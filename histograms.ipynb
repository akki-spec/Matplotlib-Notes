```python
import matplotlib.pyplot as plt
import numpy as np

#Histogram = A visual representation  of the distribution of quantitative data.
#            They grow values into bins (intervals)
#            and counts how many fall in each range.


scores = np.random.normal(loc=80,scale=50,size=1000)
plt.hist(scores)
plt.show()

```

```python


scores = np.random.normal(loc=80,scale=50,size=1000)
scores = np.clip(scores,0,100)#all scores below 0 set to 0 and above 100 set to 100

plt.hist(scores)
plt.show()
```

```python


scores = np.random.normal(loc=80,scale=50,size=1000)
scores = np.clip(scores,0,100)#all scores below 0 set to 0 and above 100 set to 100

plt.hist(scores,
         bins=10,
         color='lightgreen',
         edgecolor = 'black')

plt.title('Test Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
```
