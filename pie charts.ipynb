```python
import matplotlib.pyplot as plt
import numpy as np

#pie charts - circular chart divided into slices to show percentage of total good for viualization distribution
```

```python
categories = np.array(["Freshmen","Sophomores","Junior","Senior"])
Values = np.array([300,275,250,225])

plt.pie(Values, labels= categories)
plt.show()
```

```python
categories = np.array(["Freshmen","Sophomores","Junior","Senior"])
Values = np.array([300,275,250,225])

plt.pie(Values, labels= categories,
        autopct= "%1.f")
plt.show()
```

```python
categories = np.array(["Freshmen","Sophomores","Junior","Senior"])
Values = np.array([300,275,250,225])

plt.pie(Values, labels= categories,
        autopct= "%1.1f")
plt.show()
```

```python
categories = np.array(["Freshmen","Sophomores","Junior","Senior"])
Values = np.array([300,275,250,225])

plt.pie(Values, labels= categories,
        autopct= "%1.1f%%")
plt.show()
```

```python
categories = np.array(["Freshmen","Sophomores","Junior","Senior"])
Values = np.array([300,275,250,225])
colors = ["skyblue","blue","lightgrey","grey"]
plt.pie(Values, labels= categories,
        autopct= "%1.1f%%",
        colors = colors)
plt.show()
```

```python
categories = np.array(["Freshmen","Sophomores","Junior","Senior"])
Values = np.array([300,275,250,225])
colors = ["skyblue","blue","lightgrey","grey"]
plt.pie(Values, labels= categories,
        autopct= "%1.1f%%",
        colors = colors,
        explode = [0,0,0,0.1],
        shadow = True,
        startangle=90)# startangle rotates in 90 degree here

plt.title("college students")
plt.show()
```
