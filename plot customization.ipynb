```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([2010,2015,2020,2025])
y = np.array([20,40,100,80])

plt.plot(x,y)
plt.show()
```

```python
x = np.array([2010,2015,2020,2025])
y = np.array([20,40,100,80])

plt.plot(x,y, marker = "o")
plt.show()
```

```python
x = np.array([2010,2015,2020,2025])
y = np.array([20,40,100,80])

plt.plot(x,y, marker = "o",
         markersize = 10)
plt.show()
```

```python
x = np.array([2010,2015,2020,2025])
y = np.array([20,40,100,80])

plt.plot(x,y, marker = "o",
         ms = 10,
         markerfacecolor = "red")
plt.show()
```

```python
x = np.array([2010,2015,2020,2025])
y = np.array([20,40,100,80])

plt.plot(x,y, marker = "o",
         ms = 10,
         mfc = "#d6114c",
         markeredgecolor = "red")
plt.show()
```

```python
from matplotlib.lines import lineStyles
x = np.array([2010,2015,2020,2025])
y = np.array([20,40,100,80])

plt.plot(x,y, marker = "o",
         ms = 10,
         mfc = "#d6114c",
         mec = "#d6114c",
         linestyle = "")#lS = dashed,dotted,dashdot,solid
plt.show()
```

```python
from matplotlib.lines import lineStyles
x = np.array([2010,2015,2020,2025])
y = np.array([20,40,100,80])

plt.plot(x,y, marker = "o",
         ms = 10,
         mfc = "#d6114c",
         mec = "#d6114c",
         ls = "dashdot",
         linewidth = 5)
plt.show()
```

```python
from matplotlib.lines import lineStyles
x = np.array([2010,2015,2020,2025])
y = np.array([20,40,100,80])

plt.plot(x,y, marker = "o",
         ms = 10,
         mfc = "#d6114c",
         mec = "#d6114c",
         ls = "dashdot",
         lw = 3,
         color = "orange")
plt.show()
```

```python
from matplotlib.lines import lineStyles
x = np.array([2010,2015,2020,2025])
y1 = np.array([20,40,100,80])
y2 = np.array([22,30,110,82])

plt.plot(x,y, marker = "o",
         ms = 10,
         mfc = "#d6114c",
         mec = "#d6114c",
         ls = "dashdot",
         lw = 3,
         color = "orange")

plt.plot(x,y2)

plt.show()
```

```python
from matplotlib.lines import lineStyles
x = np.array([2010,2015,2020,2025])
y1 = np.array([20,40,100,80])
y2 = np.array([22,30,110,82])
y3 = np.array([30,70,80,3])

line_style = dict(marker = "o",
         ms = 10,
         mfc = "#d6114c",
         mec = "#d6114c",
         ls = "dashdot",
         lw = 3,
         )


plt.plot(x,y,**line_style,color = "orange")
plt.plot(x,y2,**line_style,color = "red")
plt.plot(x,y3,**line_style, color = "#2b11d6")

plt.show()
```
